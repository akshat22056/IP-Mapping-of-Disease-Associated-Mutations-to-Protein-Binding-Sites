```python
import csv
import re

# Define amino acid categories
aa_category = {
    'A': 'hydrophobic',
    'V': 'hydrophobic',
    'I': 'hydrophobic',
    'L': 'hydrophobic',
    'M': 'hydrophobic',
    'F': 'hydrophobic',
    'W': 'hydrophobic',
    'P': 'hydrophobic',  # Debatable but acceptable
    'G': 'hydrophobic',  # Debatable (ambivalent)
    'S': 'hydrophilic',
    'T': 'hydrophilic',
    'C': 'hydrophilic',  # Corrected from hydrophobic
    'N': 'hydrophilic',
    'Q': 'hydrophilic',
    'Y': 'hydrophilic',
    'D': 'acidic',
    'E': 'acidic',
    'K': 'basic',
    'R': 'basic',
    'H': 'basic'
}

#  Prepare counters for just the four conversions 
# hydrophobic  hydrophilic, acidic  basic
conversion_counts = {
    ('hydrophobic', 'hydrophilic'): 0,
    ('hydrophilic', 'hydrophobic'): 0,
    ('acidic',       'basic'):       0,
    ('basic',        'acidic'):      0,
}

same_category = 0
stop_codon    = 0
invalid_aa    = 0
invalid_format = 0

# Regex pattern to parse mutation entries
pattern = re.compile(r'^p\.([A-Z])(\d+)([A-Z*]|=)')

with open('filtered_gene_data (1).csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        mutations = row['MUTATION_AA'].split(' | ')
        for mutation in mutations:
            mutation = mutation.strip()
            match = pattern.match(mutation)
            if not match:
                invalid_format += 1
                continue
            original_aa = match.group(1)
            mutated_aa = match.group(3)

            # Handle synonymous mutations (no change)
            if mutated_aa == '=':
                mutated_aa = original_aa

            # Skip stop codons and invalid amino acids
            if original_aa not in aa_category or mutated_aa not in aa_category:
                if mutated_aa == '*' or original_aa == '*':
                    stop_codon += 1
                else:
                    invalid_aa += 1
                continue

            orig_cat = aa_category[original_aa]
            mut_cat = aa_category[mutated_aa]

            if orig_cat == mut_cat:
                same_category += 1
            else:
                key = (orig_cat, mut_cat)
                if key in conversion_counts:
                    conversion_counts[key] += 1
                # any other crosscategory is ignored

# Print the results
print("Conversion Counts Between Categories:")
for (orig, mut), count in conversion_counts.items():
    print(f"{orig:12s} -> {mut:12s}: {count}")

print("\nAdditional Stats:")
print(f"Same Category Mutations:       {same_category}")
print(f"Mutations Leading to Stop Codon: {stop_codon}")
print(f"Invalid Amino Acid Entries:     {invalid_aa}")
print(f"Invalid Format Entries:         {invalid_format}")

# Plot 1: Conversion counts
labels = [f"{o}{t}" for (o, t) in conversion_counts.keys()]
values = list(conversion_counts.values())

plt.figure()
plt.bar(labels, values)
plt.xlabel('Conversion Type')
plt.ylabel('Count')
plt.title('Amino Acid Property Conversions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 2: Additional stats
stats_labels = ['Same Category', 'Stop Codon', 'Invalid AA', 'Invalid Format']
stats_values = [same_category, stop_codon, invalid_aa, invalid_format]

plt.figure()
plt.bar(stats_labels, stats_values)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Additional Mutation Parsing Stats')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
